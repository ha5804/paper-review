# __2.Problem__

- 기존의 feature_extraction(hand-craft) 방식의 한계
- 기존의 activate function의 gradient vanishing 문제

기존 방식의 이미지 처리에서는, edge 검출과 같은 기법을 통해 feature_extraction을 하였다. 이러한 방식은 대량의 Image에 적합하지 않았고, 고해상도 이미지의 등장으로 어려움을 겪게 되었다. 또한 activate function이 CNN의 convoultion 층이 깊어 질수록, chain rule에 의해 기울기가 소실하는 문제가 발생하였다.

# __4.How Reduce Training time__

#### (1)Relu : `f(x) = max(0,x)`

- 기울기 소실 문제 : Relu함수는 미분값이 항상 1로, 기존 activation function의 기울기 소실문제를 해결하였다.
- Sparse activation : max(0, x)를 통해 중요한 feature을 제외한 뉴런을 비활성화 함으로, overfitting 방지 및 representation의 효율을 높였다.

#### (2)LRN : LRN은 normalization의 기법들중 하나이다. 

- 핵심 아이디어는 `같은 위치의 여러 feature map이 서로 경쟁하도록 만드는 것` 이다. 예를들어, edge = 8, texture = 7, color = 2...처럼 convolution layer을 통과한 data에 대해 여러 feature_map이 생성된다. 이때 edge와 texture 값을 보면 8과 7로 차이가 별로 나지 않는것을 볼 수 있다. 즉, 둘 사이의 중요성을 판단하기 위해 더 큰 feature 값에 중요도를 부여 하기 위해 사용한다. 이는 특정 feature map의 activation을 주변 feature map activation 크기로 나눠서 정규화 하는 방식으로 계산한다.

#### (3)Overlapping Pooling

- slide_window 방식으로 pooling 영역을 의도적으로 겹치게 설계하였다.
이는 더 안정적인 feature을 생성한다.
    - `stride, kernel 개념 복습`


# __5.How preventing Overfitting__

- (1)Data augmentation : `기존 데이터 변형하여 새로운 training 데이서 생성`
    - Random crop, horizontal flip, rotation, color jitter과 같은 방법론을 사용하여 데이터를 변형한다. 이러한 방식은 data가 특정 이미지를 외우는것을 방지한다. `데이터 다양성 증가(distribution 확장), invariance 학습(불필요한 정보 민감성 감소), decision boundary 정확`
    
- (2)Drop out : `학습할 때 뉴런 랜덤하게 비활성화`
    - 반복적인 학습은 training시, 특정 뉴런 조합에 의존하게 될 수 있다. 이때 drop out을 통해 뉴런을 랜덤하게 비활성화 함으로써, 일반화 성능을 높일 수 있다.
