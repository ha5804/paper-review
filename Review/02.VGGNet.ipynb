# __2.Problem__

- AlexNet의 CNN이후, depth를 깊게하기 위한 연구가 시작되었다. AlexNet의 kernel(filter)은, 11x11, 5x5, 3x3 으로, 큰 kernel에 의해 Parameter수가 많았다. 이는 깊은 네트워크를 만드는데 어려움이 있었다.

# __4. 3x3 kernel usage__

- 위 논문의 가장 핵심적인 부분으로, kernel의 크기가 커진다면 Parameter 수가 증가한다. 즉, 기존 AlexNet의 kernel크기를 줄였음에도 성능을 유지하였다. 이는 kernel을 줄임으로써 네트워크를 더 깊게 쌓는게 가능하게 되었기 때문이다.

- 예를들어 5x5 kernel이 존재한다고하면, 25개의 parameter에, 

`input -> conv(5x5) -> relu` 의 과정을 거친다.

그러나 3x3 kernel을 2번 수행한다면, 3x3x2 = 18개의 parameter에,

`input -> conv(9) -> relu -> conv(9) -> relu` 의 과정을 거친다.

이는 더 많은 활성화 함수를 통해 비선형성을 효과적으로 추가 할수 있음을 보여준다.

# __5.Discussion__
- 논문을 읽으면서, 이 논문의 핵심은 무엇인지에 대해 고민해보았다. AlexNet은 결국 확실하게 contribution이 존재하였다. Relu의 사용이 그 예이다. 그러나 VGG는 실제 구현에서도 parameter수가 AlexNet에 비해 훨씬 많았고, 그 층 또한 깊었다. 그래도 확실히 이해한건, `작은 filter 사용으로 깊은 층을 쌓으면 그 표현이 진해진다` 이다. 그러나 CNN의 근본적인 한계였던 많은 계산량과 parameter로 인해 등장한 AlexNet에서 작은 filter 사용과 깊은 네트워크로 성능을 올린것이 많은 한계와 보완해야할 점이 보인다고 생각한다.
